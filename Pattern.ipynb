```python
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact *= i
print("Factorial =", fact)
```

```python
rows = 5
for i in range(rows):
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

```python
num = 1
for i in range(1, rows + 1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))
if a >= b and a >= c:
    print("Largest:", a)
elif b >= a and b >= c:
    print("Largest:", b)
else:
    print("Largest:", c)
```

```python
rows = 5
for i in range(rows):
    print(" " * (rows - i - 1), end="")
    print("* " * (i + 1))
```

```python
rows = 5
for i in range(rows):
    print(" " * i, end="")
    print("* " * (rows - i))
```

```python
rows = 5
for i in range(rows):
    for j in range(rows):
        if i == 0 or i == rows - 1 or j == 0 or j == rows - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
rows = 5
for i in range(rows):
    ch = ord('A')
    for j in range(i + 1):
        print(chr(ch), end=" ")
        ch += 1
    print()
```
